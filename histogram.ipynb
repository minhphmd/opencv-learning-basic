```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

## Histogram

Histogram là một biểu đồ thống kê về mức độ phân phối ánh sáng trên các pixel ảnh

Tính bằng OpenCV: `calcHist(images, channelds, mask, histSize, ranges)`
* `images` - ảnh cần tính histogram
* `channels` - kênh màu dùng để thống kê
* `mask` - vùng thống kê, truyền vào "None" nếu thống kê toàn bộ ảnh
* `histSize` - số lượng bin, chia khoảng giá trị độ sáng của pixel thành `histSize` đoạn để thống kê
* `ranges` - khoảng giá trị của pixel

Vẽ biểu đồ histogram bằng matplotlib

```python
# vẽ biểu đồ historam trên ba kênh màu RGB
img = cv.imread('images/messi.jpg')
color = ('b', 'g', 'r')
for (i, channel) in enumerate(color):
    hist = cv.calcHist([img], [i], None, [256], [0, 256]) # chia đoạn [0, 256) thành 256 khoảng để thống kê histogram
    plt.plot(hist, color=channel) # plot biểu đồ ứng với kênh màu 'channel' tương ứng
plt.show()
```

```python
# vẽ biểu đồ historam về độ sáng
img = cv.imread('images/messi.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY) # chuyển ảnh thành gray scale
hist = cv.calcHist([img], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```
