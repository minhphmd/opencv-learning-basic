```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
```

## Background subtraction

Background subtraction là bài toán xác định foreground và background của ảnh thông qua một camera tĩnh. Những pixel foreground là những pixel thuộc những vật có chuyển động giữa các frame trong video. Hai phương pháp thực hiện background subtraction trong OpenCV là MOG2 và KNN. Thực hiện hai phương pháp đó thông qua lớp giải thuật `BackgroundSubtractorMOG2` và `BackgroundSubtractorKNN`. 

#### MOG2

```python
cap = cv.VideoCapture("../images/vtest.avi") # đọc video
# ret, frame = cap.read() # đọc frame đầu tiên của video

backSub = cv.createBackgroundSubtractorMOG2() # tạo lớp giải thuật MOG2

count_frame = 0

plt.rcParams['figure.dpi'] = 400

while True:
    count_frame += 1

    ret, frame = cap.read() # đọc các frame trong video

    if not ret: break

    fgMask = backSub.apply(frame)

    print(fgMask.shape)

    if count_frame % 2 == 0:
        display.display(plt.gcf())
        display.clear_output(wait=True)
        plt.subplot(1, 2, 1)
        plt.imshow(cv.cvtColor(frame, cv.COLOR_BGR2RGB))
        plt.subplot(1, 2, 2)
        plt.imshow(fgMask, cmap='gray', vmin=0, vmax=255)
        plt.show()

    if count_frame == 250:
        break
```
